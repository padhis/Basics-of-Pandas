# Calculating Summary Statistics

### Aggregating Statistics

```python
#Read the file
titanic = pd.read_csv("/content/titanic_dataset.csv")
titanic.head(5)
```

```python
#Finding the average age of the passenger
titanic["Age"].mean()
```

```python
#Find median age and ticket fare price of the Titanic passenger
titanic[["Age", "Fare"]].median()
```

# Aggregating statistics grouped by category

```python
#What is the average age for male versus female Titanic passengers?
#groupby - split, apply, combine
titanic[["Sex", "Age"]].groupby("Sex").mean()
```

```python
#An alternate approach
titanic.groupby("Sex")["Age"].mean()
```

```python
#What is the mean ticket fare price for each of the sex and cabin class combinations?
titanic.groupby(["Sex", "Pclass"])["Fare"].mean()
```

# Count number of records by category

```python
#What is the number of passengers in each of the cabin classes?
titanic["Pclass"].value_counts()
```

# How to combine data from multiple tables

## Concatenating objects

```python
air_quality_no2 = pd.read_csv("/content/air_quality_no2_long.csv", parse_dates = True)
air_quality_no2 = air_quality_no2[["date.utc", "location", "parameter", "value"]]
air_quality_no2.head(2)
```

```python
air_quality_pm25 = pd.read_csv("/content/air_quality_pm25_long.csv", parse_dates = True)
air_quality_pm25 = air_quality_pm25[["date.utc", "location", "parameter", "value"]]
air_quality_pm25.head()
```

```python
#I want to combine the measurements of no2 and pm25 two tables with a similar structure, in a single table.
air_quality = pd.concat([air_quality_no2, air_quality_pm25], axis = 0)
air_quality.head()
```

```python
print(air_quality_no2.shape)
print(air_quality_pm25.shape)
print(air_quality.shape)
```

## Join tables using a common identifier

```python
stations_coord = pd.read_csv("/content/air_quality_stations.csv")
stations_coord.head(4)
```

```python
#Add the station coordinates, provided by the stations metadata table, to the corresponding rows in the measurements table.
air_quality = pd.merge(air_quality, stations_coord, how = "left", on = "location")
air_quality.head()
```

# How to handle time series data with ease

## Using pandas datetime properties

```python
air_quality = pd.read_csv("/content/air_quality_no2_long.csv")
air_quality = air_quality.rename(columns = {"date.utc":"datetime"})
air_quality.head(3)
```

```python
#Convert the datetime column to datetime
#initially the datetime are treated as strings and do no provide any datetime operations
air_quality["datetime"] = pd.to_datetime(air_quality['datetime'])
air_quality['datetime']
```

```python
#Create a new column with only month
air_quality['month'] = air_quality['datetime'].dt.month
air_quality.head(3)
```

```python
air_quality.groupby([air_quality['datetime'].dt.weekday, "location"])["value"].mean()
```

```python

air_quality.groupby(air_quality['datetime'].dt.hour)["value"].mean().plot(kind = 'bar')
```

# How to manipulate textual data

```python
titanic = pd.read_csv("/content/titanic_dataset.csv")
titanic.head(2)
```

```python
#Make all name characters lowercase.
titanic["Name"].str.lower()
```

```python
#Create a new column Surname that contains the surname of the passengers by extracting the part before the comma
titanic['Surname'] = titanic["Name"].str.split().str.get(0)
titanic['Surname']
```

```python
#Extract the passenger data about the countesses on board of the Titanic.
titanic[titanic["Name"].str.contains("Countess")]
```

```python
#Which passenger of the Titanic has the longest name?
titanic.loc[titanic["Name"].str.len().idxmax(), "Name"]
```

```python
#In the “Sex” column, replace values of “male” by “M” and values of “female” by “F”.
titanic["sex_short"] = titanic["Sex"].replace({"male":'M', "female":'F'})
titanic
```
